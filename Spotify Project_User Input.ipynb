```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.billboard.com/charts/hot-100/"
response = requests.get(url)
response.status_code 
soup = BeautifulSoup (response.content)
```

```python
top100title = []
for title in soup.select('h3.c-title.a-no-trucate'):
    top100title.append(title.get_text(strip=True))
top100title
```

```python
top100artist = []
for artist in soup.select('span.c-label.a-no-trucate'):
    top100artist.append(artist.get_text(strip=True))
top100artist
```

```python
top100_df = pd.DataFrame({"Song": top100title, "Artist": top100artist})
top100_df
```

```python
# Save as CSV
top100_df.to_csv('top100_songs.csv', index=False)
```

```python
# Version 2 with recommendation for another song

# User Input
user_input = input("Enter a song title: ")

# User input and song titles to lowercase -> case insensitive
user_input_lower = user_input.lower()
top100_df['Song_lower'] = top100_df['Song'].str.lower()

# Check if user input is in the list
if user_input_lower in top100_df['Song_lower'].values:
    print(f"{user_input} is hot!")
    recommended_song = top100_df[top100_df['Song_lower'] != user_input_lower].sample(1)[['Song', 'Artist']].values[0]
    recommended_song_title, recommended_artist = recommended_song

    print(f"Here's another hot song for you: {recommended_song_title.capitalize()} by {recommended_artist.capitalize()}")
else:
    print(f"Sorry, the song '{user_input}' is not hot.")
    
# Drop the temporary column used for case-insensitive comparison
top100_df.drop(columns=['Song_lower'], inplace=True)
```

```python
# Version 3 with Loop to input another song

while True:
    # Input a song title
    user_input = input("Enter a song title: ")

    # User input and song titles to lowercase
    user_input_lower = user_input.lower()
    top100_df['Song_lower'] = top100_df['Song'].str.lower()

    # Check if user input is in the list
    if user_input_lower in top100_df['Song_lower'].values:
        print(f"{user_input} is hot!")

        # Recommend another song
        recommended_song = top100_df[top100_df['Song_lower'] != user_input_lower].sample(1)[['Song', 'Artist']].values[0]
        recommended_song_title, recommended_artist = recommended_song

        print(f"You might like this song as well: {recommended_song_title.capitalize()} by {recommended_artist.capitalize()}")
        
        # Drop the temporary column used for case-insensitive comparison
        top100_df.drop(columns=['Song_lower'], inplace=True)
        
        break 
    else:
        print(f"Sorry, the song '{user_input}' is not hot. Please enter another song.")

        # Drop the temporary column used for case-insensitive comparison
        top100_df.drop(columns=['Song_lower'], inplace=True)
```

```python
# Version 4

from fuzzywuzzy import fuzz

while True:
    # Input a song title
    user_input = input("Enter a song title: ")

    # User input song
    user_input_lower = user_input.lower()
    # song titles to lowercase
    top100_df['Song_lower'] = top100_df['Song'].str.lower()

    # Check if user input is an exact match or a fuzzy match
    exact_match = user_input_lower in top100_df['Song_lower'].values
    fuzzy_match = top100_df['Song_lower'].apply(lambda x: fuzz.partial_ratio(user_input_lower, x) > 80).any()

    if exact_match or fuzzy_match:
        print(f"{user_input} is hot!")

        # Recommend another song
        recommended_song = top100_df[top100_df['Song_lower'] != user_input_lower].sample(1)[['Song', 'Artist']].values[0]
        recommended_song_title, recommended_artist = recommended_song

        print(f"You might like this song as well: {recommended_song_title.capitalize()} by {recommended_artist.capitalize()}")
        
        # Drop temporary column 
        top100_df.drop(columns=['Song_lower'], inplace=True)
        
        break 
    else:
        print(f"Sorry, the song '{user_input}' is not hot. Please enter another song.")

        # Drop temporary column
        top100_df.drop(columns=['Song_lower'], inplace=True)
```
