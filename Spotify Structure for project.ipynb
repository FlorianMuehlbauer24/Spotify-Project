```python
import spotipy
import pandas as pd
import json
from spotipy.oauth2 import SpotifyClientCredentials


#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= config.client_id,
                                                           client_secret= config.client_secret))



results = sp.search(q="daddy cool",limit=5,market="GB")
results
```

### All genres playlist

```python
# 1 Get Playlist ID
# 52oPdG5WK1KcMf78FvFXIa
```

```python
# 2 Get track Ids from Playlist

def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
all_genre_tracks = get_playlist_tracks("Jonathan","52oPdG5WK1KcMf78FvFXIa")

all_genre_ids = []
for i in all_genre_tracks:
    all_genre_ids.append(i['track']['id'])
all_genre_ids 
```

```python
# Check lenght
len(all_genre_ids)
```

```python
# 3. Chop Ids Into Chunks
bp_ids = []
for i in all_genre_tracks:
    bp_ids.append(i['track']['id'])
```

```python
len(bp_ids)
```

```python
chunk_size = 100
bp_ids_chopped = [bp_ids[i:i+chunk_size] for i in range(0,len(bp_ids), 100)]
```

```python
len(bp_ids_chopped)
```

```python
#4 Get Audio Features for each chunk
af_list = []
for i in bp_ids_chopped:
    af_list.extend(sp.audio_features(i))
```

```python
af_list
```

```python
# 5. Create data frame
df_spotifysongs = pd.DataFrame(af_list)
```
